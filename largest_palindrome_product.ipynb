Find the largest palindrome from 3 digit products.

Such as the 2 digit solution: 9009, because 91 x 99 product is the largest number that can be read the same way forwards and backwards.

```python
# now unfortunately, we need to create our
# isPalindrome function here. Or else we will
# run into an error, function called before it
# is was assigned. look for "FIRST" for chronological
# steps taken
def isPalindrome(p):
    p = str(p)
    if p == p[::-1]:
        return True
    return False

# FIRST we create a list of palindromes
palindromes =  []
# creates lowest 3 digit numbers
i = 100
j = 100
# we created a double nested loop
it = 1 # iterations of j, causing less duplicates
while i < 1000:
    while j < 1000:
        # gives us our product
        product = i * j
        # if it is a palindrome
        if isPalindrome(product):
            palindromes.append(product)
        j += 1
    i += 1
    j = it + 1 # necessary to multiply all combinations

print(palindromes)
```

```python
# finds the largest number in our list
def largest_num(numbers):
    numbers.sort()
    return numbers[len(numbers) - 1]

print(largest_num(palindromes))
```
