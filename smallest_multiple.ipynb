Trying to find a more optimal solution for finding smallest multiple evenly divided by any range:

```python
# Collecting is prime from a previous problem
def isPrime(n):
    if n % 2 == 0:
        return False
    i = 3
    while i < n:
        if n % i == 0:
            return False
        i += 1
    return True

isPrime(15)
```

```python
def commonMultiplesInRange(min, max):
    primes = []
    while min <= max:
        if isPrime(min):
            primes.append(min)
        min += 1

    if primes == []:
        return 0

    answer = 1
    for i in primes:
        answer *= i

    return answer

commonMultiplesInRange(1, 10)
```

```python
# WE DID IT! in the code block below which works
# up to 28 in runtime 3.1s ... can we get 30?
# to optimize further I'm collecting primes
# here so that we can maybe adjust the function
# below to check against primes rather than
# 1-10, n*10 times
def getPrimes(min, max):
    p = []
    while min < max:
        if isPrime(min):
            p.append(min)
    return p
getPrimes(1, 30)
```

```python
def smallestMultipleV2(min, max):
    # finds the answer to half our max
    half = max / 2

    answer = max * (max - 1)

    pointer = 0
    while min <= half:
        if isPrime(min):
            min = min
            break
        min += 1

    pointer = min
    while pointer <= half:
        if isRemain(answer, pointer):
            answer += 1
            pointer = min
        pointer += 1
    print("half answer", answer)

    # finds the answer to our max
    steps = answer
    while pointer <= max:
        if isRemain(answer, pointer):
            answer += steps
            pointer = min
        pointer += 1
    return answer


smallestMultipleV2(1, 28)
```
